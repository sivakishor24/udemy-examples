```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=50
D=50
```

```python
X=(np.random.random((N, D))-0.5)*10
```

```python
#First 3 values are 1,0.5,-0.5 remaining values are 0's
ww=np.array([1,0.5,-0.5] + [0]*(D-3))
```

```python
Y=X.dot(ww) + np.random.random(N) * 0.5
```

```python
costs = []
```

```python
learning_rate = 0.001
```

```python
w = np.random.rand(D)/np.sqrt(D)
```

```python
np.random.rand(D)
```

```python
np.sqrt(D)
```

```python
w
```

```python
l1 = 10.0
```

```python
for i in range(500):
    Yhat = X.dot(w)
    delta = Yhat - Y
    w = w - learning_rate * (np.dot(X.T, delta) + l1 * np.sign(w))
    mse = np.dot(delta,delta)/N
    costs.append(mse)
```
