```python
ones = np.ones((N,1))
ones
```

```python
r = np.sqrt((X*X).sum(axis=1)).reshape(-1,1)
r
```

```python
Xb = np.concatenate((ones,r,X),axis = 1)
Xb
```

```python
Xb.shape
```

```python
#four weights for four columns
w = np.random.rand(D + 2)
w
```

```python
z = Xb.dot(w)
z
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
Y = sigmoid(z)
Y
```

```python
def cross_entropy(T, Y):
    return -(T*np.log(Y)+(1-T)*np.log(1-Y)).sum()
```

```python
cross_entropy(T, Y)
```

```python
learning_rate = 0.001
```

```python
cost = []
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range(5000):
    e = cross_entropy(T, Y)
    cost.append(e)
    if i % 500 == 0:
        print(e)
    #l2 regularization
    w = w + learning_rate*(Xb.T.dot(T-Y)-0.01*w)
    Y = sigmoid(Xb.dot(w))
```

```python
plt.plot(cost)
plt.title('Cross_entropy per iteration')
plt.show()
```

```python
print("final value:",w)
```

```python
print("final classification rate:",1 - np.abs(T - np.round(Y)).sum()/N)
```
