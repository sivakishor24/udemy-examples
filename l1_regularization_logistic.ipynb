```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
D = 50
```

```python
X = (np.random.random((N, D)) - 0.5 ) * 10
```

```python
s=np.random.random((N, D))
s
```

```python
s-0.5
```

```python
#first 3 rows are 1.0,0.5,-0.5 remaining 47 values are 0's
w1 = np.array([1.0, 0.5, -0.5] + [0] * (D-3))
w1
```

```python
len(w1)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
sigmoid(np.dot(X,w1))
```

```python
np.random.randn(N)*0.5
```

```python
Y = np.round(sigmoid(np.dot(X,w1)+np.random.rand(N)*0.5))
Y
```

```python
cost = []
learning_rate = 0.001
```

```python
w = np.random.randn(D)/np.sqrt(D)
w
```

```python
np.sqrt(D)
```

```python
l1 = 3.0
```

```python
#converting negative values to positive values
np.abs(w)
```

```python
sigmoid(X.dot(w))
```

```python
for i in range(1000):
    Ypred = sigmoid(X.dot(w))
    delta = Ypred - Y
    w = w - learning_rate * (X.dot(delta) + l1 * np.sign(w))
    c = -(Y * np.log(Ypred) + (1 - Y) * np.log (Ypred)).mean() + l1 * np.abs(w).mean()
    cost.append(c)
```

```python
cost
```

```python
plt.plot(cost)
plt.show()
```

```python
print("final values:",w)
```
