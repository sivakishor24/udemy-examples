```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=[]
Y=[]

for i in open("C:/Users/Dell/Desktop/datasets/data_poly.csv"):
    x, y = i.split(',')
    #print(x,y)
    x = float(x)
    y = float(y)
    X.append([1,x,x*x]) # x^2+x+c
    Y.append(y)
```

```python
print(X)
print(Y)
```

```python
X=np.array(X)
Y=np.array(Y)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,1],Y)
plt.show()
```

```python
#calculate weights
w=np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat=np.dot(X,w)
print(w)
```

```python
print(Yhat)
```

```python
print(Yhat.size)
print(Yhat.dtype)
```

```python
total=pd.DataFrame({'Actual values':Y, 'Predicted values':Yhat})
print(total.head())
```

```python
plt.scatter(X[:,1],Y)
plt.plot(sorted(X[:,1]),sorted(Yhat))
plt.show()
```

```python
#compute  r-squared
d1=Y-Yhat
d2=Y-Y.mean()
print("data for d1:",d1)
print("data for d2:",d2)
```

```python
#r2_score
r2=1-d1.dot(d1)/d2.dot(d2)
print("r-squared:",r2)
```

```python
from sklearn.metrics import r2_score
print("r2_score is :",r2_score(Y,Yhat))
```
