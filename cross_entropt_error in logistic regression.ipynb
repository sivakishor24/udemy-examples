```python
N= 100
D = 2
```

```python
X = np.random.rand(N,D)
```

```python
#first 50 rows are subtracted that by negative 2
#center the first 50 points at(-2,-2)
X[:50, :] = X[:50, :]-2*np.ones((50, D))
```

```python
X[:50, :]
```

```python
#last 50 rows are added with positive 2 values
#last 50 points at(2,2)
X[50:, :] = X[50:, :]+2*np.ones((50, D))
```

```python
X[50:, :]
```

```python
#we have to set first 50 rows are 0's and last 50 rows are 1's
#label the first 50 are 0's and last 50 are 1's
T = np.array([0]*50 + [1]*50)
T
```

```python
#we have to creat bias terms
#add a column of ones
ones = np.array([[1]*N]).T
ones
```

```python
#we have to concatenate the X values and bias values
Xb = np.concatenate((ones, X), axis = 1)
Xb
```

```python
#we have to calculate the weights
#randomly initilize the weights
w = np.random.rand(D + 1)
w
```

```python
#calculate the prediction output
z = np.dot(Xb,w)
z
```

```python
#another method
z= Xb.dot(w)
z
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
Y = sigmoid(z)
Y
```

```python
#calculate the cross entropy error
#So we can take in the target and the predicted output 
def cross_entropy(T, Y):
    E = 0
    for i in range(N):
        if T[i] == 1:  # here target value is 1
           E -= np.log(Y) #here subtract log Y
        else:       #here target value is 0
           E -= np.log(1 - Y[i])
    return E
```

```python
print(cross_entropy(T, Y))
```

```python
#here weights depends only on the means
#here zero is the bias term and 4 is weights
w = np.array([0, 4, 4])

z = Xb.dot(w)
z
```

```python
Y = sigmoid(z)
Y
```

```python
print(cross_entropy(T, Y))
```
