```python
N = 100
D = 2
```

```python
X = np.random.randn(N,D)
print(X)
```

```python
X[:50, :] = X[:50, :] - 2*np.ones((50,D))
X
```

```python
X[50:,:] = X[50:, :] + 2*np.ones((50, D))
X
```

```python
T = np.array([0]*50 + [1]*50)
print(T)
```

```python
ones = np.ones((N, 1))
ones
```

```python
Xb = np.concatenate((X, ones), axis = 1)
Xb
```

```python
w = np.random.randn(D + 1)
w
```

```python
z = Xb.dot(w)
z
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
Y = sigmoid(z)
Y
```

```python
def cross_entropy(T, Y):
    E = 0
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1-Y[i])
    return E
```
