```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
#Load the data
X=[]
Y=[]
for line in open("C:/Users/Dell/Desktop/datasets/data_2d.csv"):
    x1,x2,y=line.split(',')
    X.append([1, float(x1), float(x2)])
    Y.append(float(y))
    
    
#turn  X and Y into numpy arrays
X=np.array(X)
Y=np.array(Y)

print(X.shape)
print(Y.shape)
```

```python
#letsplot the data to see the data looks like
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```

```python
#calculate weights
w=np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat=np.dot(X,w)
print(w)
```

```python
print(Yhat)
```

```python
import pandas as pd
compare=pd.DataFrame({'Actual values':Y, 'Predicted values':Yhat})
print(compare.head())
```

```python
#compute  r-squared
d1=Y-Yhat
d2=Y-Y.mean()
r2=1-d1.dot(d1)/d2.dot(d2)
print("r-squared:",r2)
```

```python
from sklearn.metrics import r2_score
print("r2_score is:",r2_score(Y,Yhat))
```
